## Symbol Maps

```python
year = '2015'
df = pd.read_csv(year + ".csv")
df['text'] = df['Country'] + '<br>Happiness Score ' + (df['Happiness Score']).astype(str)
df.head()
```

```python
## Merge happiness data with geographical location
```

```python
country_latlon = pd.DataFrame(index=range(245),columns=range(3))
country_latlon.iloc[0]
with open('country_latlon.csv', encoding="latin-1") as datafile:
    i = 0
    for line in datafile:
        if i != 0:
            country_latlon.iloc[i - 1] =line.split(',')[1], line.split(',')[2], line.split(',')[3][:-1]
        i += 1
        if i == 246:
            break
country_latlon.columns = ['latitude', 'longitude', 'country']
country_latlon.tail()
```

```python
country_latlon.iloc[2,2]
```

```python
merge_2015 = pd.merge(df, right=country_latlon, how='left', left_on='Country', right_on='country')
```

```python
import numpy as np
df = merge_2015
limits = [(0,30),(31,60),(61,90),(91,120),(121,157)]
# colors = ['#ffffe0','#ffd59b','#ffa474','#f47461','#db4551','#b81b34','#8b0000']
colors = ["#fde725ff", "#5cc863ff", "#21908dff", "#3b518bff", "#440154ff"]
countries = []
scale = 0.4
for i in range(len(limits)):
    lim = limits[i]
    df_sub = df[lim[0]:lim[1]]
    country = dict(
        type = 'scattergeo',
        lon = df_sub['longitude'],
        lat = df_sub['latitude'],
        text = df_sub['text'],
        marker = dict(
            size = np.power(df_sub['Happiness Score']/scale,2),
            color = colors[i],
            line = dict(width=0.5, color='rgb(40,40,40)'),
            sizemode = 'area'
        ),
        name = '{0} - {1}'.format(lim[0],lim[1]) )
    countries.append(country)
    
layout = dict(
        title = '2014 US city populations<br>(Click legend to toggle traces)',
        showlegend = True,
        geo = dict(
            scope="world",
            showland = True,
            landcolor = 'rgb(217, 217, 217)',
            subunitwidth=1,
            countrywidth=1,
            subunitcolor="rgb(255, 255, 255)",
            countrycolor="rgb(255, 255, 255)"
        ),
    )

fig = dict( data=countries, layout=layout )
iplot( fig, filename='d3-bubble-map-populations' )
```
